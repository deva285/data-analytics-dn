# **7,Print list of Top N states by the least number of death in a specific year. Also needed to be ranke**d

## **A,Top 10 least death states in 2020**


```python
def get_year(row):
    return datetime.datetime.strptime(row['date'], '%Y-%m-%d').year
df3['year'] = df3.apply(get_year, axis=1)
df_2020 = df3.where(df3['year'] == 2020)
df_top10_2020 = df_2020.groupby(['state'])['deaths'].sum().to_frame('sum_deaths').reset_index()
df_top10_2020 = df_top10_2020.sort_values(by=['sum_deaths'], ascending=True)

def new_rank(row):
    return int( 56 - row['rank1'])
df_top10_2020['rank1'] = df_top10_2020['sum_deaths'].rank(ascending = False )
df_top10_2020['rank'] = df_top10_2020.apply(new_rank, axis=1)
df_top10_2020_sel = df_top10_2020[['state', 'sum_deaths', 'rank']]
df_top10_2020_sel.head(10)
```

## **B,Top 10 least death states in 2021**

## **a.Top 10 least death states in last three months of the 2021**

```python
def get_year(row):
    return datetime.datetime.strptime(row['date'], '%Y-%m-%d').year
def get_month(row):
    return datetime.datetime.strptime(row['date'], '%Y-%m-%d').month
df3['year'] = df3.apply(get_year, axis=1)
df3['month'] = df3.apply(get_month, axis=1)
df_2021 = df3.where(df3['year'] == 2021)
df_2021_3months = df_2021.where( (df_2021['month'] == 10) | (df_2021['month'] == 11) | (df_2021['month'] == 12))
df_top10_2021_3months = df_2021_3months.groupby(['state'])['deaths'].sum().to_frame('total_deaths_2021').reset_index()
df_top10_2021_3months = df_top10_2021_3months.sort_values(by=['total_deaths_2021'], ascending=True)
def new_rank(row):
    return int( 56 - row['rank1'])
df_top10_2021_3months['rank1'] = df_top10_2021_3months['total_deaths_2021'].rank(ascending = False )
df_top10_2021_3months['rank'] = df_top10_2021_3months.apply(new_rank, axis=1)
df_top10_2021_3months_sel = df_top10_2021_3months[['state', 'total_deaths_2021', 'rank']]
df_top10_2021_3months_sel.head(10)


from google.colab import drive
drive.mount('/content/drive')
```
